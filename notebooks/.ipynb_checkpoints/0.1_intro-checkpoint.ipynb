```python
def get_binary(num, digits=5):
    """get the 0-filled binary representation"""
    return (bin(num)[2:]).zfill(digits)
```

```python
for i in range(32):
    key = get_binary(i) 
    
    # decrypt with key
    break
```

```python
def decrypt(message, key):
    """decrypt the message with key
    """
    return "".join([decrypt_one(m, k) for m, k in zip(message, key)])
```

```python
def decrypt_one(message, key):
    """Decrpyt one letter with key
    """
    message, key = as_gf2(message), as_gf2(key)
    return str(message + key  )
```

```python
def as_gf2(word):
    """string as gf2 object"""
    if word in ['0', 0]:
        return 0
    if word in ['1', 1]:
        return one
```

```python
from string import ascii_lowercase
alphabet = [letter for letter in ascii_lowercase] + [' ']

def to_alphabet(message: str):
    """string message to alphabet"""
    num = int(message, 2)
    return alphabet[num]
```

```python
def alphabetical_message(message, key):
    return "".join([to_alphabet(decrypt(m, key)) for m in message])
```

```python
for i in range(32):
    key = get_binary(i) 
    
    # decrypt with key
    try:
        print(f"{i=}, {key=}, {alphabetical_message(message, key)=}")
    except Exception:
        pass
```

```python
def my_filter(L, num):
    """
    Args:
        L: list
        num: positive int
    
    Returns:
        list of numbers not containing a multiple of num
    """
    return [a for a in L if a % num != 0]
```

```python
my_filter([1, 2, 4, 5, 7], 2)
```

```python
def my_lists(L):
    """
    Args:
        L: list of non-negaive integers
    
    Returns:
        a list of lists, for every element x in L create a list containing 1, 2, ...x
        
    e.g,
        [1, 2, 4] -> [[1], [1, 2], [1, 2, 4]]
        [0] -> [[]]
    """
    L.sort()
    all_lists = []
    for i in range(len(L)):
        all_lists.append([a for a in L[:i+1] if a > 0])
    return all_lists

def my_lists_2(L): return [[a for a in sorted(L)[:i+1] if a > 0] for i in range(len(L))]
```

```python
my_lists([0])
```

```python
my_lists([1, 2, 4])
```

```python
my_lists_2([1, 2, 4])
```

```python
my_lists_2([4, 1, 2, 2])
```

```python
def my_function_composition(f, g):
    """
    Args:
        f,g: 2 functions represented as dictionaries, such that g o f exists
    Returns:
        dictionary that represents g o f
        
    f: {0: 'a', 1: 'b'}, g: {'a': 'apple', 'b': 'banana'} -> {0: 'apple', 1: 'banana'}
    """
    return {key: g[value] for key, value in f.items()}
```

```python
my_function_composition({0: 'a', 1: 'b'}, {'a': 'apple', 'b': 'banana'})
```

```python
def mySum(L):
    """
    Args:
        L: list of numbers
    Returns:
        sum of numbers in the list
    """
    current = 0
    for x in L:
        current += x
    return current
```

```python
mySum([2, 4, 5])
```

```python
def myProduct(L):
    current = 1
    for x in L:
        current *= x
    return current
```

```python
myProduct([3, 4, 3])
```

```python
def myMin(L):
    current = np.inf
    for x in L:
        if x < current:
            current = x
    return current
```

```python
myMin([3, 4, 321])
```

```python
def myConcat(L):
    current = ""
    for x in L:
        current += str(x)
    return current
```

```python
myConcat([3, 3, 4, ])
```

```python
def myUnion(L):
    current = set()
    for x in L:
        current.update(x)
    return current
```

```python
myUnion([{3, 4}, {5, 4}])
```
